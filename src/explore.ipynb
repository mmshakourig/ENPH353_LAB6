```python
def plateSegmentor(in_path, plate_name) -> list:

        plate = cv2.imread(in_path + plate_name)

        pl_ch1 = plate[80:240,:,2]
        pl_binary = pl_ch1 > 100
        
        out1 = pl_binary[:, 0:150]
        out2 = pl_binary[:, 150:300]
        out3 = pl_binary[:, 300:450]
        out4 = pl_binary[:, 450:600]
        
        pl_name = plate_name.split(".")[0]

        l_char1 = LabelledCharacter(out1, pl_name[-4])
        l_char2 = LabelledCharacter(out2, pl_name[-3])
        l_char3 = LabelledCharacter(out3, pl_name[-2])
        l_char4 = LabelledCharacter(out4, pl_name[-1])

        return [l_char1, l_char2, l_char3, l_char4]

class LabelledCharacter:

    def __init__(self, image, letter) -> None:
        self.image = image
        # self.label = label
        self.letter = letter

    def getImage(self) -> np.ndarray:
        return self.image
    
    def getLabel(self) -> np.ndarray:
        return None
    
    def getLetter(self) -> str:
        return self.letter
```
